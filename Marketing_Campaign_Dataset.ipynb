# Distribution of customer education level

```python
education_level_distribution = df['Education'].value_counts()
education_level_distribution
```

```python
plt.figure(figsize = (10,6))
df['Education'].value_counts().plot(kind = 'bar', color = 'lightcoral')
plt.title("Education Level Distribution")
plt.xlabel('Education Level')
plt.ylabel('Number of Customers')
plt.show()
```

# What is the age distribution of customers?


```python
plt.figure(figsize = (10,6))
df['Year_Birth'].plot(kind = 'hist', bins = 20, edgecolor = 'black', color = 'red')
plt.title("Age Distribution of Customers")
plt.xlabel('Year of birth')
plt.ylabel('Frequency')
plt.show()
```

# What are the most frequently purchased products (wines, fruits, meat)

```python
df.info()
```

```python
product = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']

total = df[product].sum()
plt.figure(figsize = (10,6))
total.plot(kind = 'bar', color = 'salmon')
plt.title("Total Spending on Products")
plt.xlabel('Product')
plt.ylabel('Total Spending')
plt.show()
```

# How does the rate of spending the products differences between age groups?

```python
df['Age'] = 2024 - df['Year_Birth']
age_bins = [20, 30, 40, 50, 60, 70, 80, 90]

df['Age_Group'] = pd.cut(df['Age'], bins = age_bins)

spend = df.groupby('Age_Group')[product].sum()
spend.plot(kind = 'bar', stacked = True, colormap = 'tab20', figsize = (14,8))
plt.title('Spending on Products by Age group')
plt.xlabel('Age Group')
plt.ylabel('Averge Spending')
plt.show()
```

```python
#removing stacked = true

df['Age'] = 2024 - df['Year_Birth']
age_bins = [20, 30, 40, 50, 60, 70, 80, 90]

df['Age_Group'] = pd.cut(df['Age'], bins = age_bins)

spend = df.groupby('Age_Group')[product].sum()
spend.plot(kind = 'bar', colormap = 'tab20', figsize = (14,8))
plt.title('Spending on Products by Age group')
plt.xlabel('Age Group')
plt.ylabel('Averge Spending')
plt.show()
```

# What is the rate of purchasing online different from purchasing in stores?

```python
df['Total_Online'] = df['NumWebPurchases'] + df['NumCatalogPurchases'] #making a new df of total online purchase
df['Total_Store'] = df['NumDealsPurchases']   #making a new variable of total store purchase

online_store = df[['Total_Online', 'Total_Store']].sum()

plt.figure(figsize = (10,6))
online_store.plot(kind = 'bar', color = ['Red', 'Blue'])
plt.title('Online vs Store Purchases')
plt.xlabel('Purchase Method')
plt.ylabel('Total Purchases')
plt.show()
```

# Checking the relationship between a customer's income and the number of online purchases from the customer.

```python
plt.figure(figsize = (10,6))
plt.scatter(df['Income'], df['Total_Online'], alpha = 0.5)
plt.title('Income vs Online Purchases')
plt.xlabel('Income')
plt.ylabel('Online Purchases')
plt.show()
```

# What are the factors that affect customer acceptance of different promotional offers?

```python
plt.figure(figsize = (10,6))
sns.boxplot(x = 'Response', y = 'Income', data = df)
plt.title('Income vs Response')
plt.xlabel('Response')
plt.ylabel('Income')
plt.show()

plt.figure(figsize = (10,6))
sns.boxplot(x = 'Response', y = 'Age', data = df)
plt.title('Age Distribution by Response')
plt.xlabel('Response')
plt.ylabel('Age')
plt.show()
```

# What factors affect customer's complain?

```python
plt.figure(figsize = (10,6))
sns.boxplot(x = 'Complain', y = 'Income', data = df)
plt.title('Complain by Income')
plt.xlabel('Complain')
plt.ylabel('Income')
plt.show()


plt.figure(figsize = (10,6))
sns.boxplot(x = 'Complain', y = 'Age', data = df)
plt.title('Complain by Age')
plt.xlabel('Complain')
plt.ylabel('Age')
plt.show()
```

# Dt_Customer baata oldest customer and newest customer ko enrollment date check garne

```python
# checking the enrollment date of newest and oldest customer from Dt_Customer

#converting it into date_time_format

df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], format = '%d-%m-%Y')

oldest_customer = df['Dt_Customer'].min()
newest_customer = df['Dt_Customer'].max()

print(f"The oldest customer enrolled on {oldest_customer}")
print(f"The newest customer enrolled on {newest_customer}")
```
